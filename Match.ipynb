```python

bangun_datar= """Masukkan pilihan :
1. Menghitung luas persegi
2. menghitung lingkaran
3. menghitung segitiga
masukan pilihan :
"""

b = input(bangun_datar)

match b:
    case "1":
        c= int(input("Berapa Sisinya:"))
        d= c*c
        print(d)
    case "2":
        c= int(input("Berapa Jari-jarinya:"))
        d= 2 * 3.14 * c
        print(d)
    case "3":
        c= int(input("Berapa alasnya?"))
        d= int(input("Berapa tingginya?"))
        e= 0.5 * c * d
        print(e)
```
